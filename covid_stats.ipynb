```python
# https://ourworldindata.org/coronavirus-source-data
df = pd.read_csv('owid-covid-data.csv')
df['cfr'] = df.total_deaths / df.total_cases * 100
```

```python
plt.style.use('seaborn')
pivot = pd.pivot_table(data=df,columns='location',index='date',values='total_cases')
NA = ['Canada','United States','Mexico']
world = ['United States','Italy','Germany','United Kingdom','South Korea','China',
        'France','Canada','Australia','Iran']
cc = ['United States','Italy','China','United Kingdom','Canada','South Korea','Australia']
colors=['blue','green','red','darkslategrey','darkmagenta','black','orange']
pivot[cc].plot.line(figsize=(11,9),fontsize=14,rot=90,logy=True,linewidth=1.75,color=colors)
plt.title('CUMULATIVE CASE GROWTH',fontweight='bold',fontsize=14)
plt.legend(loc='upper right',bbox_to_anchor=[1.3,1],fontsize=15)
```
